Question:15-->>>How does Python's garbage collection system work?

Answer:15-->

--> Python's garbage collection system automatically removes objects from memory that are no longer being used.

---> It helps free up space so that the program doesn't waste memory or slow down over time.

--->>> Python mainly uses a technique called reference counting—every object keeps track of how many times it’s being used. When that count becomes zero, it means no one needs it, so it gets deleted.

--> On top of that, Python also has a cyclic garbage collector to handle special cases where two objects refer to each other but are not used anywhere else.

--->> So, we don’t have to delete things manually—Python’s garbage collector keeps cleaning things behind the scenes to make our program run smoothly.

```python
# Question:17--->> Write a Python program to create and write a list of numbers to a file, one number per line?
# Answer:17-->>>

# Let's say Sameer wants to save some numbers to a file

numbers = [5, 10, 15, 20, 25]

with open("sameer_numbers.txt", "w") as file:
    for num in numbers:
        file.write(str(num) + "\n")  # writing each number on a new line

print("Numbers written to 'sameer_numbers.txt' one per line.")

#This will give output: Numbers written to 'sameer_numbers.txt' one per line.
#And the file will have:
# 5
# 10
# 15
# 20
# 25
```

```python
# Question:18--->> How would you implement a basic logging setup that logs to a file with rotation after 1MB?
# Answer:18-->>>

import logging
from logging.handlers import RotatingFileHandler

# Set up a rotating log handler
log_file = "pwskills_rotation.log"
handler = RotatingFileHandler(log_file, maxBytes=1 * 1024 * 1024, backupCount=3)

# Configure logging
logging.basicConfig(
    level=logging.INFO,
    handlers=[handler],
    format="%(asctime)s - %(levelname)s - %(message)s"
)

# Write multiple logs to test rotation
for i in range(10000):
    logging.info(f"Logging entry number {i} by Sameer for rotation testing")

print("Logging with rotation has started. Logs will rotate after 1MB.")

#This will give output: Logging with rotation has started. Logs will rotate after 1MB.
#Once log size reaches 1MB, it will create backup files like:
# pwskills_rotation.log.1, pwskills_rotation.log.2 etc.
```

```python
# Question:19--->> Write a program that handles both IndexError and KeyError using a try-except block?
# Answer:19-->>>

# We'll try to access a list and a dictionary wrongly to raise both errors

try:
    marks = [88, 76, 92]
    print(marks[5])  # This will raise IndexError

    student = {"name": "Sameer", "age": 21}
    print(student["grade"])  # This will raise KeyError if the above line is removed
except IndexError:
    print("Caught IndexError: Tried to access an index that doesn't exist in the list.")
except KeyError:
    print("Caught KeyError: Tried to access a key that doesn't exist in the dictionary.")

#This will give output: Caught IndexError: Tried to access an index that doesn't exist in the list.
#If we comment the IndexError line, then KeyError will be triggered and handled instead.
```

```python
# Question:20--->> How would you open a file and read its contents using a context manager in Python?
# Answer:20-->>>

# We'll use 'with' statement which is the context manager in Python

with open("pwskills_notes.txt", "r") as file:
    content = file.read()
    print("File content:\n", content)

#This will give output: It will print the contents of 'pwskills_notes.txt'
#Using 'with' ensures the file is closed automatically after reading.
```

```python
# Question:21--->> Write a Python program that reads a file and prints the number of occurrences of a specific word?
# Answer:21-->>>

# Let's say Sameer wants to count how many times the word "python" appears in his notes file

word_to_count = "python"

try:
    with open("sameer_notes.txt", "r") as file:
        content = file.read().lower()  # converting to lowercase for accurate match
        count = content.count(word_to_count.lower())
        print(f"The word '{word_to_count}' occurred {count} times in the file.")
except FileNotFoundError:
    print("File not found. Please check if 'sameer_notes.txt' exists.")

#This will give output like: The word 'python' occurred 3 times in the file.
#If file is not found → it will show a proper message.
```

```python
# Question:22--->> How can you check if a file is empty before attempting to read its contents?
# Answer:22-->>>

# Let's check if Sameer's file has any data before reading it

import os

file_path = "sameer_data.txt"

if os.path.exists(file_path):
    if os.stat(file_path).st_size == 0:
        print("The file is empty.")
    else:
        with open(file_path, "r") as file:
            content = file.read()
            print("File content:\n", content)
else:
    print("The file does not exist.")

#This will give output:
# - If file is empty → The file is empty.
# - If file has content → it will print the content.
# - If file doesn't exist → it will show proper message.
```

```python
# Question:23--->> Write a Python program that writes to a log file when an error occurs during file handling?
# Answer:23-->>>

import logging

# Setting up basic logging
logging.basicConfig(filename="file_error_log.txt", level=logging.ERROR,
                    format='%(asctime)s - %(levelname)s - %(message)s')

try:
    with open("missing_file_by_sameer.txt", "r") as file:
        data = file.read()
        print("File content:\n", data)
except FileNotFoundError as e:
    logging.error(f"File not found error occurred: {e}")
    print("Oops! File not found. Error logged.")

#This will give output:
# → Oops! File not found. Error logged.
# And in 'file_error_log.txt' the error will be saved with timestamp and message.
```
